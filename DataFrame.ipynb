# Merge, Join and Concatenate : 

- left
- right
- inner
- outer


```python
import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\onkar.dapake\OneDrive - PROWESS SELLING SKILLS PRIVATE LIMITED\Prowess Gen AI\Courses\dataframe1.csv")
df1
```

```python
df2=pd.read_csv(r"C:\Users\onkar.dapake\OneDrive - PROWESS SELLING SKILLS PRIVATE LIMITED\Prowess Gen AI\Courses\dataframe2.csv")
df2
```

```python
df1.merge(df2) #inner join : only matching data from both sheet
```

```python
df1.merge(df2, how = 'inner', on = ['ID'])   # how inner join takes or apply it shows
```

```python
df1.merge(df2, how = 'left') # left join
```

```python
df1.merge(df2, how = 'outer') # outer join
```

```python
df1.merge(df2, how = 'cross') # cross join
```

# Join

```python
df1.join(df2,on = 'ID', how = 'outer', lsuffix= '_Left', rsuffix= '_Right')
```

```python
df4 = df1.set_index('ID').join(df2.set_index('ID'), lsuffix= '_Left', rsuffix= '_Right')
df4
```

# Concatenate: 

```python
pd.concat([df1,df2])
```

```python
df1.append(df2)
```
